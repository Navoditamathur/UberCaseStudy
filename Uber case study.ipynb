The time taken for each trip can be calculated as

```python
df['Time_taken']=df['Drop_timestamp']-df['Request_timestamp']
```

Now divide the Requests according to the day times
1.Morning
2.Afternoon
3.Evening
4.Night
5.Late night

```python
df['Request_timestamp']=df['Request_timestamp'].astype(str)
```

```python
df.loc[:, 'Request_date'] = df['Request_timestamp'].apply(lambda x: x.split(" ")[0])


df.loc[:, 'Request_time'] = df['Request_timestamp'].apply(lambda x: x.split(" ")[1])
```

```python
df.head()
```

```python
df['Request_timestamp']=pd.to_datetime(df['Request_timestamp'])
df['Drop_timestamp']=pd.to_datetime(df['Drop_timestamp'])
bins = [0, 5, 11, 17, 19, 24]
labels = ['Late night','Morning','Afternoon','Evening','Night']
hours = df['Request_timestamp'].dt.hour
df['bin'] = pd.cut(hours,bins=bins,labels=labels,right=False)
```

```python
df
```

```python

plt.figure(figsize=(16, 14))

plt.subplot(2, 1, 1)
p = sns.countplot(x='bin', data=df)
p.set_xticklabels(p.get_xticklabels(),rotation=30)
plt.title('Frequency of requests accaording to bin')

plt.subplot(2, 1, 2)
q = sns.countplot(x='bin', hue='Status', data=df)
q.set_xticklabels(q.get_xticklabels(),rotation=30)
plt.title('Frequency of requests accaording to bin that shows cancelled or no cars available')


plt.show()
```
